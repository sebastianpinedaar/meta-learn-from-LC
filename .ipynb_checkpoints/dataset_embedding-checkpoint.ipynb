```python
data_inverse = {}
X_complete = []
y_complete = [] 
y_first_time = []
best_conf_per_dataset = {}
matrix = np.zeros((len(data.keys()), len(data[list(data.keys())[0]]["x_first"])))

for d, dataset in enumerate(data.keys()):
    best_conf_per_dataset[dataset] = {"perf":0, "conf":"None"}
    for i, x in enumerate(data[dataset]["x_first"]):  
        config = "".join([str(hp) for hp in x[0]])
        
        if config not in data_inverse:
            data_inverse[config] = {"X": [], "y_val":[], "dataset_name":[]}
        X_complete.append(x[0]+data[dataset]["meta_feat"])
        #perf = x[2]/(x[1]+1e-10)
        perf = x[2]
        y_first_time.append(x[1])
        y_complete.append((x[2],x[1]))
        data_inverse[config]["X"].append(data[dataset]["meta_feat"])
        #data_inverse[config]["y_val"].append(x[2]/(x[1]+1e-10))
        data_inverse[config]["y_val"].append(perf)
        data_inverse[config]["dataset_name"].append(dataset)
        
        if perf> best_conf_per_dataset[dataset]["perf"]:
            best_conf_per_dataset[dataset]["perf"] = perf
            best_conf_per_dataset[dataset]["conf"] = config
            
        matrix[d, i] = x[1]
```

```python
from torch import nn
lr = 0.005
epochs = 50
ranker = MLP(29, n_hidden=20, n_layers=1, n_output = 1, use_cnn =False).to(device)   

optimizer = torch.optim.Adam(ranker.parameters(), lr= lr)


loss_fn = nn.MarginRankingLoss(margin=0.01)


losses = [np.inf]
batch_size = 100
for epoch in range(epochs):
    optimizer.zero_grad()
    
    with torch.no_grad():
        batch_id = torch.randint(0, X.shape[0], (batch_size,))
        X_batch = X[batch_id]
        y_batch = y[batch_id]
        
        idx = torch.arange(batch_size)
        id_cross = torch.cartesian_prod(idx, idx)
        target = y_batch[id_cross[:,0]] >= y_batch[id_cross[:,1]]
        target = target.long()
        target[target==0] = -1
        target[id_cross[:,0]==id_cross[:,1]]=0
    
    pred = ranker(X_batch)
    loss = loss_fn(pred[id_cross[:,0]], pred[id_cross[:,1]], target)
    loss.backward()
    optimizer.step()
    losses.append(loss.detach().cpu().numpy().item())
```
